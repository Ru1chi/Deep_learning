## **Resizing the Image**

```python
from PIL import Image#pillow librabry used and has a module as Image

```

```python
img=Image.open("/content/dog-img.jpg")
img_resized=img.resize((200,200))#give in 2 brackets
```

```python
img_resized.save('dog_image_resized.jpg')
```

```python
image_res=mpimg.imread("/content/dog_image_resized.jpg")
img_res_plot=plt.imshow(image_res)
```

## **Convert RGB image to Grayscale image using OpenCV**

```python
import cv2
```

```python
img=cv2.imread("/content/dog-img.jpg")
```

```python
type(img)
```

```python
#convert to grayscale image
img.shape
```

```python
grayscale_img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)#pass 2 arguments ,the 1 is the image and the other is RGB2GRAY
```

```python
type(grayscale_img)
```

```python
grayscale_img
```

```python
grayscale_img.shape
```
