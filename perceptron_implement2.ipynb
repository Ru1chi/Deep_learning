```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("/content/diabetes.csv")
```

```python
df.head()
```

```python
df['Outcome'].value_counts()#there is an imbalance in the dataset
```

```python
#undersampling the class 0(non-diabetic) to get equal samples
class_0_df=df[df['Outcome']==0]
class_1_df=df[df['Outcome']==1]

class_0_df=class_0_df.sample(268)#sample function randomly sample the datapoints

data=pd.concat([class_0_df,class_1_df])
```

```python
#now seperate input features and target features
x=data.drop('Outcome',axis=1)
y=data['Outcome']
```

```python
#as the input features are in pandas dataFrame but we need it in numpy array
x=x.values
y=y.values
```

```python
#now perform train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
scaler=StandardScaler()
```

```python
x_train_sclaed=scaler.fit_transform(x_train)
```

```python

x_test_scaled=scaler.transform(x_test)
```

```python
#sigmoid Perceptron
perceptron=sigmoidPerceptron(input_size=x_train_sclaed.shape[1])#x_trained_scleadis used for the number of columns
```

```python
#train the perceptron model
perceptron.fit(inputs=x_train_sclaed,targets=y_train,learning_rate=0.1,num_epochs=100)
```

```python
#model evaluation for test data
accuracy=perceptron.evaluate(x_train_sclaed,y_train)
accuracy=round(accuracy*100,2)
print('Training Accuracy:',accuracy,'%')
```

```python
#model evaluation for training data
accuracy=perceptron.evaluate(x_test_scaled,y_test)
accuracy=round(accuracy*100,2)
print('Testing Accuracy:',accuracy,'%')
```
